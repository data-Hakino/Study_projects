```python
df = df.rename(columns = {'Номер' : 'number', 
               'Дата создания' : 'create_date', 
               'Дата оплаты' : 'payment_date', 
               'Title' : 'title', 
               'Статус' : 'status',
               'Заработано' : 'money', 
               'Город' : 'city', 
               'Платежная система': 'payment_system'})
```

```python
df.groupby('title', as_index=False)\
.aggregate({'money': 'sum'})\
.sort_values('money', ascending=False)
#перенос строки одной команды. Улучшает читабельность кода.
```

```python
# В качестве теоретической задачи было узнать сколько каждый продукт заработал в каждом городе
# Поэтому группировку нужно выполнить также и по городам
df_city_money = df.groupby(['title', 'city'], as_index=False)\
                .aggregate({'money': 'sum'})\
                .sort_values('money', ascending=False)
```

```python
df_city_money
```

```python
# Полученный datafreim можно преобразовать в отдельный файл. 
# Алгоритм: берем переменную, в скобках указываем её название и отключаем индексы.
df_city_money.to_csv('df_city_money.csv', index = False)
# Файл сохранен и повявился в той же директории, где находится файл этого ноутбука
```

```python
# Аналог where в SQL это функция query. В скобках подается требуемое нам условие.
df\
.query('status == "Завершен"')\
.groupby('title', as_index=False)\
.aggregate({'money': 'sum', 'number':'count'})\
.sort_values('money', ascending=False)
#перенос строки одной команды. Улучшает читабельность кода.
#ДОЛГО НЕ МОГ ПОНЯТЬ ПОЧЕМУ У МЕНЯ ТУТ НЕ РАБОТАЕТ КОД. Оказалось, что я при переименовании столбцов я случайно поставил пробел перед number -_-
```

```python
# Поскольку в переменной number теперь хранится количество УСПЕШНО заверешенных заказов, то будет грамотно её переименовать
money_title = df\
.query('status == "Завершен"')\
.groupby('title', as_index=False)\
.aggregate({'money': 'sum', 'number':'count'})\
.sort_values('money', ascending=False)\
.rename(columns=({'number': 'success'}))
```

```python
# После формирования выборки(отчета) логично сделать проверку достоверности данных.
# Так как в отчет включены только успешные заказы, которые оплачены, то сумма денег полученная от курсов в данном отчете, должна быть равна сумме
# средств полученных в исходной таблице df
money_title.money.sum()
```

```python
all_money = df.money.sum()
all_money
```

```python
# Можно реализовать проверку через блок if/else
if int(money_title.money.sum()) == int(all_money):
    print('OK')
else:
    print('Беда-беда')
```

```python
# библиотека datedime. Команда для получения сегодняшней даты
today_date = datetime.today().strftime('%Y-%m-%d')
today_date
```

```python
file_name = 'money_title {}.csv'
```

```python
# Автоматическая вставка в название файла текущей даты
file_name = file_name.format(today_date)
```

```python
# При успешной проверке данных до и после агрегации будет записан файл money_title с актуальной датой
if int(money_title.money.sum()) == int(all_money):
    print('OK')
    money_title.to_csv(file_name, index = False)
else:
    print('Беда-беда')
```
